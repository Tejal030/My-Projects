```python
import time 
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

# Global variables

TIME = 4 # time in seconds. This is how long selenium will run to load the page. If selenium isn't loading the page fast enough, then you can increase this value

#!!! DO NOT REMOVE THE r IN THE FRONT of the PATH variable !!!
# if you are using Windows and selenium isn't picking up the PATH properly,
# then visit this page:  https://clay-atlas.com/us/blog/2019/10/27/python-english-tutorial-solved-unicodeescape-error-escape-syntaxerror/
PATH = r"/Users/tejalambilwade/Desktop/XN_Project/chromedriver"

URL = "https://www.udacity.com/course/c-plus-plus-nanodegree--nd213"

def get_soup_dynamic(url):
    
    # open up connection and grabbing the page
    ser = Service(PATH)
    op = webdriver.ChromeOptions()
    browser = webdriver.Chrome(service=ser, options=op)
    
    browser.get(url)
    time.sleep(TIME)
    html = browser.page_source
    page_soup = soup(html, "html.parser")
    return page_soup

#Fetching the page
Page = get_soup_dynamic(URL)

#Print Page Title
page_title = Page.title.text
print("\n================================TITLE=======================================\n")
print(page_title)

#Getting the course description
direct_path_to_course_description = "#main > div > div.rc-XdpSection.cdp-about > div > div > div > div._1b7vhsnq.m-t-2 > div.m-t-1.description > div"
course_description = Page.select(direct_path_to_course_description)[0].get_text()
print("\n==========================COURSE DESCRIPTION================================\n")
print(course_description)

#Getting the instructor list
direct_path_to_instructor_list = "#main > div > div.rc-XdpSection.cdp-instructors > div > div"
instructor_list = Page.select(direct_path_to_instructor_list)[0].get_text()
print("\n============================INSTRUCTOR LIST==================================\n")
print(instructor_list)

#Getting the syllabus
direct_path_to_syllabus = "#main > div > div.rc-XdpSection.cdp-syllabus.border-top.border-bottom > div > div > div:nth-child(2)"
syllabus = Page.select(direct_path_to_syllabus)[0].get_text()
print("\n================================SYLLABUS=====================================\n")
print(syllabus)

#Getting the course reviews
direct_path_to_reviews = "#reviews > div > div._10hnekkx > div"
reviews = Page.select(direct_path_to_reviews)[0].get_text()
print("\n================================REVIEWS======================================\n")
print(reviews)

#Getting all related courses
related_courses = [element.text for element in Page.find_all("div", {"class": "m-a-1s"})]
print("\n============================RELATED COURSES==================================\n")
print(related_courses)

#Getting all enrollment options
direct_path_to_enrollment_options = "#main > div > div:nth-child(11) > div > div > div > div > div"
enrollment_options = Page.select(direct_path_to_enrollment_options)[0].get_text()
print("\n===========================ENROLLMENT OPTIONS================================\n")
print(enrollment_options)

#Getting all FAQ Questions and Answers
direct_path_to_faq = "#main > div > div.rc-XdpSection.cdp-faqs > div > div > div > div > ul"
faq = Page.select(direct_path_to_faq)[0].get_text()
print("\n==================================FAQ=========================================\n")
print(faq)


## CAN YOU THINK OF A BETTER WAY TO GRAB BOTH QUESTIONS AND ANSWERS AT THE SAME TIME COUPLE THEM TOGETHER ?
## ALSO TRY TO GRAB THE UNIVERSITIES IN THE "ONLINE DEGREES SECTION"
```
